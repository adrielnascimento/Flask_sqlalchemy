```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# gerar 10 valores que vai de 0 ate 5
t = np.linspace(0, 20*np.pi, 100)
```

```python
y = np.cos(t)
```

```python
# grafico com titulos e legendas nos eixos
plt.plot(t, y)
plt.title('grafico do cosseno')
plt.xlabel('tempo (s)')
plt.ylabel('amplitude ')
plt.show()
```

```python
plt.plot(t, y, color='red', linestyle='--', marker=0, label='dados 1')
plt.xlabel('tempo (S)')
plt.ylabel('cosseno')
plt.title('grafico do cosseno')
plt.legend()
plt.show()
```

```python
# grafico de barras
categorias = ["A", "B", "C", "D"]
valores = [15, 20, 10, 25]
plt.bar(categorias, valores, color='red')
plt.xlabel('categorias')
plt.ylabel('valores')
plt.title('grafico de barras')
plt.show()
```

```python
#grafico de disperso
plt.figure('disperso')
plt.scatter(categorias, valores, color='green', marker='x')
plt.grid()
plt.show()
```

```python
#histograma
dados = np.random.normal(0, 1, 1000)
plt.hist(dados, bins=20, edgecolor='black')
plt.show()
```

```python
# cosseno
plt.figure('cosseno')
plt.plot(t, y, marker='o', color='red', label='dados 2')
plt.xlabel('tempo (s)')
plt.ylabel('cosseno')
plt.title('grafico cosseno')
plt.legend()
plt.show()
```

```python
# seno
y1 = np.sin(t)
plt.figure('seno', figsize=(5, 4)) # figsize altera o tamanho da imagem
plt.plot(t, y1, marker='o', color='green', label='dados 2')
plt.title("grfico do seno")
plt.ylabel('seno')
plt.xlabel('tempo (S)')
plt.legend()
plt.show()
```

```python
anos = [2015, 2016, 2017, 2018, 2019]
vendas = [500, 600, 750, 800, 950]
# 2 graficos dentro da mesma figura
plt.figure(figsize=(12, 5))
# 1 grafico
plt.subplot(1, 2, 1)
plt.plot(anos, vendas, marker='o', color='blue')
plt.title('vendas ao longo dos anos')
# grafico 2
plt.subplot(1, 2, 2)
plt.plot(categorias, valores, marker='o', color='green')
plt.title('vendas por categoria')
plt.show()
```

```python
a = np.arange(0, 100)
b = np.exp(a)
plt.figure('grafico', figsize=(5, 4))
plt.plot(a, b,color='blue', label='exp')
plt.show()
```

```python
a = np.linspace(0, 2 * np.pi, 100)
sen = np.sin(a)
cos = np.cos(a)
plt.figure('comparao', figsize=(8, 6))
plt.plot(sen, color='blue', label='seno')
plt.plot(cos, color='green', label='cosseno')
plt.title('comparao seno cosseno')
plt.legend()
plt.grid()
plt.show()
```

```python
# subplot
x = np.linspace(0, 2*np.pi, 500)
c = np.cos(x)
s = np.sin(x)
```

```python
plt.figure('subplots', figsize=(12, 6))
# grafico 1
axis1 = plt.subplot(1, 2, 1) # linha 1, coluna 2, posio 1
axis1.set_title('grafico cosseno')
axis1.plot(x, c, label='cosseno')
axis1.set_xlabel('tempo')
axis1.set_ylabel('longetude')
axis1.legend()
axis1.grid()

#grafico 2
axis2 = plt.subplot(1, 2, 2) # linha 1, coluna 2, posio 2
axis2.set_title('grafico seno')
axis2.plot(x, s, label='seno')
axis2.set_xlabel('tempo')
axis2.set_ylabel('longetude')
axis2.legend()
axis2.grid()
plt.show()
```

```python
a = np.linspace(0, 10, 100)
y1 = np.cos(a)
y2 = np.sin(a)
y3 = np.tan(a)
y4 = np.exp(-a)

fig, ax = plt.subplots(2, 2, figsize=(12, 6))

#grfico do cosseno
ax[0, 0].plot(a, y1, color='blue')
ax[0, 0].set_title('grfico cosseno')
ax[0, 0].set_xlabel('tempo')
ax[0, 0].grid()

#grafico seno
ax[0, 1].plot(a, y2, color='green')
ax[0, 1].set_title('grfico seno')
ax[0, 1].set_xlabel('tempo')
ax[0, 1].grid()

#grafico tangente
ax[1, 0].plot(a, y3, color='green')
ax[1, 0].set_title('grafico da tangente')
ax[1, 0].set_xlabel('tempo')
ax[1, 0].grid()

# exponencial decrescesnte
ax[1, 1].plot(a, y4, color='purple')
ax[1, 1].set_title('grafico exponencial')
ax[1, 1].set_xlabel('tempo')
ax[1, 1].grid()

#ajustar os espaos do grficos
plt.tight_layout()
plt.show()
```
